# Load tips datasets from seaborn

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

## Get all rows with smoker = no and total_bill >= 10

```python
smoker = tips.loc[(tips['smoker'] == 'No') & (tips['total_bill'] >= 10)]
smoker
```

```python
### get avg of total_bil for each smoker,d
```

```python
avg = tips.groupby(['smoker', 'day','time'])['total_bill'].mean()
avg
```
