```python
import pandas as pd
```

```python
billboard = pd.read_csv('../scipy-2019-pandas/data/billboard.csv')
billboard.tail()
```

```python
(billboard
     .melt(
         id_vars=['year','artist','time','track','date.entered'], 
         var_name='weeks',
         value_name='rank',
     )
     .groupby('artist')['rank']
     .mean()
     .sort_values(ascending=False)
     .head(20)
     
)
```
