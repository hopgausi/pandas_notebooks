```python
import pandas as pd
```

```python
ebola = pd.read_csv('../scipy-2019-pandas/data/country_timeseries.csv')
ebola.head()
```

```python
ebola_df=ebola.melt(id_vars=['Date','Day'],var_name='status_country',value_name='count')
ebola_df.head()
```

```python
status_country = ebola_df['status_country'].str.split('_', expand=True)
status_country.head() 
```

```python
ebola_df[['Status','Country']] = status_country
ebola_df.head()
```

```python
ebola_df.drop(['status_country'],axis=1, inplace=True)
ebola_df.head()
```
